Ok, vamos a desmenuzarlo paso a paso para que quede claro por qué esa integral

$$
\theta = \int_0^1 g(x) \, dx
$$

puede escribirse como

$$
\theta = E[g(U)]
$$

cuando $U \sim \text{Uniforme}(0,1)$.

---

## 1️⃣ Definición de esperanza para variables continuas

Si $X$ es una variable aleatoria continua con densidad $f_X(x)$, entonces:

$$
E[g(X)] = \int_{-\infty}^{\infty} g(x) \, f_X(x) \, dx
$$

---

## 2️⃣ Caso especial: uniforme en (0,1)

Si $U \sim \text{Uniforme}(0,1)$, entonces su función de densidad es:

$$
f_U(u) = 1 \quad \text{para } 0 < u < 1
$$

y $f_U(u) = 0$ fuera de ese intervalo.

---

## 3️⃣ Sustituyendo en la fórmula de la esperanza

Si calculamos $E[g(U)]$:

$$
E[g(U)] = \int_{0}^{1} g(u) \cdot f_U(u) \, du
$$

Pero como $f_U(u) = 1$ en todo el intervalo:

$$
E[g(U)] = \int_{0}^{1} g(u) \cdot 1 \, du
$$

$$
E[g(U)] = \int_{0}^{1} g(u) \, du
$$

Esto es exactamente $\theta$.

---

## 4️⃣ Interpretación intuitiva

* La integral $\int_0^1 g(x) dx$ es el “área bajo la curva” de $g(x)$ en $(0,1)$.
* El valor esperado $E[g(U)]$ es el promedio de $g(X)$ cuando $X$ se elige al azar de manera uniforme entre 0 y 1.
* La uniformidad de $U$ hace que cada punto del intervalo tenga el mismo peso en el promedio, así que el promedio *es* el área normalizada de la curva.

---

## 5️⃣ Conexión con Monte Carlo

Este vínculo es clave: si quieres calcular $\theta$ pero no puedes integrar $g(x)$ de forma analítica, puedes:

1. Generar $n$ valores $u_1, u_2, \dots, u_n$ \~ Uniforme(0,1).
2. Calcular $g(u_i)$ para cada uno.
3. Usar:

$$
\theta \approx \frac{1}{n} \sum_{i=1}^n g(u_i)
$$

Con $n$ grande, por la **Ley de los Grandes Números**, esta media se acerca a la integral.
