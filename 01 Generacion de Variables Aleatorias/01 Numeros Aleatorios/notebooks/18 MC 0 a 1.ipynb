# **MONTE CARLO EN [0,1]**

Perfecto, lo resolvemos **a mano** paso a paso usando Monte Carlo.

---

## **1. Planteamiento del problema**

Queremos aproximar:

$$
\theta = \int_0^1 e^{e^x} \, dx
$$

Usando el m√©todo de Monte Carlo, interpretamos la integral como:

$$
\theta = E\big[ g(U) \big], \quad U \sim \text{Uniform}(0,1), \quad g(u) = e^{e^u}
$$

La idea es generar $n$ valores $U_i$ uniformes en $(0,1)$ y estimar:

$$
\hat{\theta}_n = \frac{1}{n} \sum_{i=1}^n g(U_i)
$$

---

## **2. Datos proporcionados**

N√∫meros pseudoaleatorios $U_i$:

$$
0.7284,\quad 0.1539,\quad 0.5872,\quad 0.9126,\quad 0.3467
$$

Tenemos $n = 5$.

---

## **3. Evaluaci√≥n de $g(u) = e^{e^u}$**

Calculemos cada t√©rmino a mano (usando 4 decimales intermedios):

1. **Para** $u_1 = 0.7284$
   $e^{u_1} \approx e^{0.7284} \approx 2.0715$
   $g(u_1) \approx e^{2.0715} \approx 7.9399$

2. **Para** $u_2 = 0.1539$
   $e^{u_2} \approx e^{0.1539} \approx 1.1664$
   $g(u_2) \approx e^{1.1664} \approx 3.2109$

3. **Para** $u_3 = 0.5872$
   $e^{u_3} \approx e^{0.5872} \approx 1.7991$
   $g(u_3) \approx e^{1.7991} \approx 6.0424$

4. **Para** $u_4 = 0.9126$
   $e^{u_4} \approx e^{0.9126} \approx 2.4909$
   $g(u_4) \approx e^{2.4909} \approx 12.0731$

5. **Para** $u_5 = 0.3467$
   $e^{u_5} \approx e^{0.3467} \approx 1.4142$
   $g(u_5) \approx e^{1.4142} \approx 4.1132$

---

## **4. Promedio muestral**

Sumamos los valores:

$$
S = 7.9399 + 3.2109 + 6.0424 + 12.0731 + 4.1132 \approx 33.3795
$$

Calculamos la estimaci√≥n:

$$
\hat{\theta}_5 = \frac{33.3795}{5} \approx 6.6759
$$

---

## **5. Interpretaci√≥n**

* **Estimaci√≥n Monte Carlo:** $\theta \approx 6.6759$ usando 5 puntos.
* Esto es una aproximaci√≥n con alta variabilidad por el tama√±o de muestra peque√±o.
* Si aumentamos $n$, la **Ley de los Grandes N√∫meros** asegura que $\hat{\theta}_n \to \theta$.
* El **error est√°ndar** se reducir√≠a como $\sigma / \sqrt{n}$.

## **1. Datos**

Valores de $g(U)$ calculados antes:

$$
g_1 = 7.9399, \quad
g_2 = 3.2109, \quad
g_3 = 6.0424, \quad
g_4 = 12.0731, \quad
g_5 = 4.1132
$$

Media muestral:

$$
\bar{g} = \hat{\theta}_5 \approx 6.6759
$$

---

## **2. Varianza muestral**

La f√≥rmula de la **varianza muestral** (no sesgada) es:

$$
s^2 = \frac{1}{n-1} \sum_{i=1}^n (g_i - \bar{g})^2
$$

Calculamos cada desviaci√≥n al cuadrado:

1. $(7.9399 - 6.6759)^2 \approx (1.2640)^2 \approx 1.5977$
2. $(3.2109 - 6.6759)^2 \approx (-3.4650)^2 \approx 12.0031$
3. $(6.0424 - 6.6759)^2 \approx (-0.6335)^2 \approx 0.4013$
4. $(12.0731 - 6.6759)^2 \approx (5.3972)^2 \approx 29.1440$
5. $(4.1132 - 6.6759)^2 \approx (-2.5627)^2 \approx 6.5683$

Suma:

$$
\Sigma \approx 1.5977 + 12.0031 + 0.4013 + 29.1440 + 6.5683 \approx 49.7144
$$

Dividimos entre $n-1 = 4$:

$$
s^2 \approx \frac{49.7144}{4} \approx 12.4286
$$

Por lo tanto:

$$
s \approx \sqrt{12.4286} \approx 3.5254
$$

---

## **3. Error est√°ndar**

El error est√°ndar del estimador Monte Carlo es:

$$
\text{SE} = \frac{s}{\sqrt{n}} \approx \frac{3.5254}{\sqrt{5}} \approx \frac{3.5254}{2.2361} \approx 1.5764
$$

---

## **4. Intervalo de confianza al 95%**

Usamos la distribuci√≥n t de Student con $n-1 = 4$ grados de libertad:

$$
t_{0.975,4} \approx 2.776
$$

El intervalo es:

$$
\bar{g} \pm t \cdot \text{SE} \quad \Rightarrow \quad
6.6759 \pm 2.776 \times 1.5764
$$

Margen de error:

$$
ME \approx 2.776 \times 1.5764 \approx 4.3778
$$

---

**Resultado final:**

$$
IC_{95\%} \approx (6.6759 - 4.3778, \; 6.6759 + 4.3778) \approx (2.2981, \; 11.0537)
$$

---

## **5. Interpretaci√≥n formal**

* El intervalo $(2.30, 11.05)$ **no significa** que la probabilidad de que $\theta$ est√© ah√≠ sea 95%, sino que **si repiti√©ramos el experimento muchas veces**, el 95% de esos intervalos contendr√≠an el valor real $\theta$.
* El **error est√°ndar** ($\approx 1.5764$) cuantifica la dispersi√≥n esperada del estimador $\hat{\theta}_n$ alrededor de $\theta$ para muestras de tama√±o 5.
* El intervalo es **muy ancho** porque $n=5$ es muy peque√±o y la funci√≥n $e^{e^x}$ crece r√°pidamente, generando alta varianza.

# **t student**

Se usa la **t de Student** y no directamente la normal por una raz√≥n clave:

---

## 1Ô∏è‚É£ Contexto

Cuando aplicamos la f√≥rmula del intervalo de confianza para una media:

$$
\bar{x} \pm z_{\alpha/2} \frac{\sigma}{\sqrt{n}}
$$

* $z_{\alpha/2}$ viene de la distribuci√≥n normal est√°ndar.
* Esto **solo es v√°lido** si conocemos la desviaci√≥n est√°ndar poblacional $\sigma$.

En Monte Carlo, igual que en la mayor√≠a de aplicaciones reales, **no conocemos $\sigma$**:

* Tenemos que estimarla con la **desviaci√≥n est√°ndar muestral** $s$.
* Esto introduce **incertidumbre extra** en el estimador, especialmente cuando $n$ es peque√±o.

---

## 2Ô∏è‚É£ Por qu√© entra la t-Student

William Sealy Gosset (bajo el pseud√≥nimo "Student") demostr√≥ que:

Si $X_1, \dots, X_n$ son i.i.d. normales con media $\mu$ y desviaci√≥n est√°ndar $\sigma$ desconocida, entonces:

$$
T = \frac{\bar{X} - \mu}{s / \sqrt{n}} \quad \text{sigue una distribuci√≥n t de Student con } n-1 \text{ grados de libertad.}
$$

üîπ La **t-Student** es m√°s "gorda" en las colas que la normal.
üîπ Esto compensa el hecho de que **usar $s$** en vez de $\sigma$ genera m√°s variabilidad.

---

## 3Ô∏è‚É£ Regla pr√°ctica

* **n grande** (ej. $n > 30$): la t se aproxima a la normal y se puede usar $z$.
* **n peque√±o** y $\sigma$ desconocida: usar $t$ con $n-1$ grados de libertad.

En tu ejercicio:

* $n = 5$ ‚Üí muy peque√±o.
* $\sigma$ no conocida ‚Üí usamos $s$ y por tanto **t-Student**.

---

## 4Ô∏è‚É£ Interpretaci√≥n

En tu intervalo $(2.30, 11.05)$, el factor $t_{0.975,4} \approx 2.776$ es **m√°s grande** que el equivalente normal $z_{0.975} \approx 1.96$.
Esto **ensancha el intervalo** para reflejar la mayor incertidumbre en la estimaci√≥n de la varianza.

## 1Ô∏è‚É£ Normal est√°ndar $N(0,1)$

* **Se usa cuando:**

  * La variable poblacional es normal **y** la desviaci√≥n est√°ndar $\sigma$ es conocida.
  * O bien, $n$ es grande ($n > 30$), $\sigma$ desconocida pero reemplazada por $s$, aplicando aproximaci√≥n asint√≥tica (Teorema Central del L√≠mite).
* **Aplicaciones t√≠picas:**

  * Intervalos de confianza para medias con $\sigma$ conocida.
  * Pruebas z para medias o proporciones.

---

## 2Ô∏è‚É£ t de Student

* **Se usa cuando:**

  * La poblaci√≥n es normal (o $n$ no muy grande) y $\sigma$ es desconocida.
  * Intervalos de confianza y pruebas de hip√≥tesis sobre medias.
* **Regla clave:**

  * Grados de libertad $\text{gl} = n - 1$.
  * Para $n$ grande, t ‚Üí normal est√°ndar.
* **Aplicaciones:**

  * Intervalos para media con varianza desconocida y $n$ peque√±o.
  * Pruebas t de una muestra, dos muestras independientes o apareadas.

---

## 3Ô∏è‚É£ Chi-cuadrado $\chi^2$

* **Se usa cuando:**

  * Se trabaja con varianzas y desviaciones est√°ndar.
  * Se modelan variables que son suma de cuadrados de normales est√°ndar.
* **Aplicaciones:**

  * Intervalos de confianza para la varianza.
  * Pruebas de bondad de ajuste.
  * Pruebas de independencia en tablas de contingencia.

---

## 4Ô∏è‚É£ F de Fisher-Snedecor

* **Se usa cuando:**

  * Se comparan **varianzas** de dos poblaciones.
  * Es el cociente de dos variables $\chi^2$ escaladas por sus grados de libertad.
* **Aplicaciones:**

  * ANOVA (an√°lisis de varianza).
  * Pruebas de homogeneidad de varianzas.

---

## 5Ô∏è‚É£ Distribuciones no centrales y casos especiales

* **Distribuci√≥n normal no central:** aparece en pruebas de hip√≥tesis con medias desplazadas.
* **t no central:** en potencias de prueba para t-tests.
* **F no central:** en ANOVA cuando la hip√≥tesis nula es falsa.

---

üí° **Resumen pr√°ctico de decisi√≥n:**

1. **¬øEstimamos una media?**

   * $\sigma$ conocida ‚Üí Normal.
   * $\sigma$ desconocida:

     * $n$ peque√±o ‚Üí t-Student.
     * $n$ grande ‚Üí Normal (aproximaci√≥n).
2. **¬øEstimamos una varianza?** ‚Üí Chi-cuadrado.
3. **¬øComparamos dos varianzas?** ‚Üí F de Fisher.
4. **¬øComparamos varias medias?** ‚Üí F de Fisher (ANOVA).
