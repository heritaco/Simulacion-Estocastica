**1) Hipergeométrica multivariante** $\mathbf X\sim\mathrm{MH}(N;K_1,\dots,K_m;n)$
Muestreo sin reemplazo con $m$ clases.

$$
\Pr(\mathbf X=\mathbf x)=\frac{\prod_{i=1}^m \binom{K_i}{x_i}}{\binom{N}{n}},\quad \sum x_i=n.
$$

Uso: manos de cartas por palo, inspección de lotes por tipo de defecto.

**2) Hipergeométrica no central (sesgo por “odds”)**
Dos formulaciones no equivalentes:

* **Fisher’s** $\mathrm{FNCH}$: condiciona en $n$. Para 2 colores,

$$
\Pr(X=k)=\frac{\binom{K_1}{k}\binom{K_2}{n-k}\,\omega^{k}}
{\sum_{j}\binom{K_1}{j}\binom{K_2}{n-j}\,\omega^{j}},
$$

$\omega>0$ es la razón de odds del color 1.
Uso: muestreo sin reemplazo con **preferencias fijas** a nivel de muestra (p. ej., unidades con distinta “atractividad”).

* **Wallenius’** $\mathrm{WNCH}$: selección **secuencial** con probabilidad proporcional a pesos restantes; la pmf se define vía integral.
  Uso: sesgo actuando **en cada extracción** (competencia continua por ser elegido).

(Versiones multivariantes existen: $\Pr(\mathbf x)\propto \prod_i \binom{K_i}{x_i}\omega_i^{x_i}$ en Fisher; en Wallenius se pondera paso a paso.)

**3) Hipergeométrica negativa** $\mathrm{NH}$
Sin reemplazo **hasta** observar $r$ éxitos. Si $X$ es el número de extracciones totales:

$$
\Pr(X=x)=\frac{\binom{x-1}{r-1}\binom{N-x}{K-r}}{\binom{N}{K}},\quad r\le x\le N-K+r.
$$

Uso: “¿cuántas piezas debo revisar hasta hallar $r$ defectuosas?” Límite $N\to\infty$ $\Rightarrow$ negativa binomial.

**4) Versiones con “contagio” o heterogeneidad**

* **Pólya urn $\Rightarrow$** conteos $\sim$ **Dirichlet–multinomial**; caso $m=2$: **beta–binomial**

$$
\Pr(\mathbf x)=\frac{n!}{\prod x_i!}\frac{B(\boldsymbol{\alpha}+\mathbf x)}{B(\boldsymbol{\alpha})}.
$$

Uso: **sobre-dispersión** por dependencias o $p$ aleatorio entre ensayos; análogo “con reemplazo reforzado”.

**Mapa mental rápido**

* Sin reemplazo, clases múltiples, sin sesgo → **MH**.
* Sin reemplazo, con sesgo: fija en la muestra → **Fisher**; por extracción → **Wallenius**.
* Sin reemplazo, regla de paro en $r$ éxitos → **NH**.
* Con “aprendizaje/contagio” o heterogeneidad de $p$ → **Dirichlet–multinomial / beta–binomial**.
* Si $n/N\to 0$: MH $\to$ multinomial; NH $\to$ negativa binomial; FNCH/ WNCH $\to$ modelos sesgados “tipo binomial”.
