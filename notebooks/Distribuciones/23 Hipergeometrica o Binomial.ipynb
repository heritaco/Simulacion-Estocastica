# Diferencia entre Hipergeométrica y Binomial

Hipergeométrica: muestreo **sin reemplazo** de una población finita.
Binomial: **ensayos independientes** con probabilidad fija $p$ (equivalente a muestreo **con reemplazo** o población “infinita”).

**Modelos y pmf**

* **Binomial** $X\sim\mathrm{Bin}(n,p)$: $\displaystyle \Pr[X=k]=\binom{n}{k}p^{k}(1-p)^{n-k}$.
* **Hipergeométrica** $X\sim\mathrm{Hiper}(N,K,n)$: $\displaystyle \Pr[X=k]=\frac{\binom{K}{k}\binom{N-K}{\,n-k\,}}{\binom{N}{n}}$.
  Aquí $N$=población, $K$=“éxitos” en la población, $n$=tamaño de muestra.

**Momentos**

* Binomial: $\mathbb E[X]=np,\ \mathrm{Var}[X]=np(1-p)$.
* Hipergeométrica: $\mathbb E[X]=n\frac{K}{N},\ \mathrm{Var}[X]=n\frac{K}{N}\Big(1-\frac{K}{N}\Big)\frac{N-n}{N-1}$.
  La razón $\frac{N-n}{N-1}$ es la **corrección por población finita**.

**Cuándo usar**

* Usa **hipergeométrica** si extraes sin reemplazo y el **fracción de muestreo** $n/N$ no es despreciable.
* Usa **binomial** si hay reemplazo, los ensayos son i.i.d. con $p$ constante, o si $n/N$ es pequeña. Regla práctica: si $n/N\le 0.1$, la binomial es una buena aproximación con $p=K/N$.

**Ejemplos**

* Cartas: $N=52$, $K=13$ corazones, robas $n=5$ **sin** reemplazo → hipergeométrica.
  Si robas **con** reemplazo o $N$ muy grande y $n$ pequeña → binomial con $p=13/52$.

**Observación límite**
$\mathrm{Hiper}(N,K,n)\ \xrightarrow[N\ \text{grande},\ n/N\to 0]{}\ \mathrm{Bin}\big(n,p=K/N\big)$.

## Ejemplos por caso.

**Usa hipergeométrica** $(N,K,n)$ sin reemplazo:

* Baraja: $N=52$, $K=13$ corazones, robas $n=5$ cartas.
* Lote finito: $N=1{,}000$ piezas, $K=40$ defectuosas, inspección $n=60$.
* Urna escolar: $N=300$ alumnos, $K=120$ mujeres, seleccionas $n=50$ para encuesta sin reposición.
* Sorteo de rifas: $N=500$ boletos, $K=5$ premiados, extraes $n=3$ ganadores sin devolver.
* Auditoría de inventario: $N=2{,}000$ ítems, $K=80$ vencidos, muestrea $n=200$.
* Muestras biológicas limitadas: $N=96$ pocillos, $K=15$ positivos, pipeteas $n=10$ sin reponer.
* Draft deportivo con bolsa fija: $N=60$ prospectos, $K=8$ pitchers, escoges $n=5$ sin reemplazo.

**Usa binomial** $(n,p)$ con reemplazo o i.i.d.:

* Lanzamientos de moneda: $n=20$, $p=0.5$.
* Ensayos A/B en web: $n$ usuarios independientes, conversión $p$ constante.
* Producción en flujo grande: cada pieza pasa prueba con prob. $p$; líneas largas ⇒ independencia razonable.
* Llamadas de venta: $n$ intentos, éxito por llamada $p$ estable.
* Control de calidad por *stream*: inspeccionas $n$ unidades de una producción “infinita”, $p$ defecto.
* PCR con réplicas independientes: $n$ corridas, éxito $p$ por réplica, sin interacción entre réplicas.
* Tiros al aro: $n$ tiros, prob. de acierto $p$ estable y sin fatiga relevante.

**Ambigua: decide por protocolo**:

* Cartas con devolución tras cada robo → binomial $(n,p=13/52)$; sin devolución → hipergeométrica $(N=52,K=13,n)$.
* Encuesta en ciudad grande: si muestreo sistemático con reposición implícita o $n/N\le 0.1$ → binomial; si se sin reposición y $n/N$ grande → hipergeométrica.

**Regla práctica**: si $n/N\le 0.1$, la binomial aproxima bien a la hipergeométrica con $p=K/N$.
