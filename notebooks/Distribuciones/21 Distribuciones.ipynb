# Notas de clase — Distribuciones: para qué sirven

> Idea núcleo: una **distribución** describe la ley generadora de un fenómeno aleatorio. Elige soporte, parámetros y una “historia mecánica” mínima que la hace operacional.

---

## Taxonomía funcional rápida

| Distribución          | ¿Qué modela?                                      |                    Soporte | Parámetros                    | Historia mecánica mínima                                                        |
| --------------------- | ------------------------------------------------- | -------------------------: | ----------------------------- | ------------------------------------------------------------------------------- |
| **Uniforme** $U(a,b)$ | Ignorancia informada, muestreo sin preferencia    |                    $[a,b]$ | $a,b$                         | Simetría y equiprobabilidad; base para **transformada inversa**.                |
| **Bernoulli**         | Éxito/fracaso de una sola prueba                  |                  $\{0,1\}$ | $p$                           | Un ensayo con prob. $p$.                                                        |
| **Binomial**          | Nº de éxitos en $n$ ensayos i.i.d.                |            $\{0,\dots,n\}$ | $n,p$                         | Suma de $n$ Bernoulli. Aproxima Poisson si $n\to\infty$, $p\to0$, $np=\lambda$. |
| **Geométrica**        | Espera hasta el **primer** éxito                  |            $\{1,2,\dots\}$ | $p$                           | Ensayos independientes hasta éxito.                                             |
| **Negativa binomial** | Espera hasta **r**-ésimo éxito                    |          $\{r,r+1,\dots\}$ | $r,p$                         | Suma de $r$ geométricas.                                                        |
| **Poisson**           | **Conteos** de eventos en tiempo/espacio con tasa |          $\{0,1,2,\dots\}$ | $\lambda$                     | Límite de binomiales raros; procesos de llegadas.                               |
| **Exponencial**       | **Tiempos entre** eventos Poisson                 |               $[0,\infty)$ | $\lambda$                     | Sin memoria. Mínimo de exponentiales es exponencial.                            |
| **Gamma**             | Suma de $r$ tiempos exponenciales                 |               $[0,\infty)$ | $k,\theta$ o $(\alpha,\beta)$ | Tiempo al $r$-ésimo evento. Conjuga a Poisson.                                  |
| **Normal**            | Sumas promedio de muchos efectos pequeños         |                $\mathbb R$ | $\mu,\sigma^2$                | TCL; ruido agregado.                                                            |
| **Lognormal**         | Productos multiplicativos de factores positivos   |               $(0,\infty)$ | $\mu,\sigma^2$                | Si $\log X\sim \mathcal N$ entonces $X$ lognormal.                              |
| **Laplace**           | Errores con colas más pesadas que normal          |                $\mathbb R$ | $\mu,b$                       | Diferencia de exponentiales i.i.d.; robustez $L^1$.                             |
| **Cauchy**            | Colas pesadas sin media/varianza                  |                $\mathbb R$ | $x_0,\gamma$                  | Cociente de normales. No aplica TCL clásico.                                    |
| **Chi-cuadrado**      | Suma de cuadrados de normales                     |               $[0,\infty)$ | $k$                           | Varianzas muestrales; bases de pruebas.                                         |
| **t de Student**      | Medias con varianza desconocida                   |                $\mathbb R$ | $\nu$                         | Cociente Normal/raíz de Chi²; colas pesadas.                                    |
| **F**                 | Cociente de varianzas                             |               $[0,\infty)$ | $\nu_1,\nu_2$                 | ANOVA, comparaciones de varianza.                                               |
| **Beta**              | Proporciones y probabilidades a priori            |                    $[0,1]$ | $\alpha,\beta$                | Conjugada de Bernoulli/Binomial.                                                |
| **Dirichlet**         | Composiciones en simplex                          |             $\Delta^{K-1}$ | $\boldsymbol\alpha$           | Conjugada de Multinomial.                                                       |
| **Categorical**       | Una etiqueta en $K$ clases                        |            $\{1,\dots,K\}$ | $\boldsymbol\pi$              | Un solo ensayo con $K$ categorías.                                              |
| **Multinomial**       | Conteos por categoría en $n$ ensayos              | $\mathbb N^K$ con suma $n$ | $n,\boldsymbol\pi$            | $n$ Categorical i.i.d.                                                          |
| **Hipergeométrica**   | Éxitos sin reemplazo                              |            $\{0,\dots,n\}$ | $N,K,n$                       | Muestra sin reemplazo de población finita.                                      |
| **Weibull**           | Vida útil y fatiga                                |               $[0,\infty)$ | $k,\lambda$                   | Tasa de fallo crec./decr.; generaliza exponencial.                              |
| **Pareto**            | Colas gruesas de riqueza/riesgo                   |             $[x_m,\infty)$ | $x_m,\alpha$                  | Procesos con invariancia de escala.                                             |
| **GPD**               | Excesos sobre umbral                              |               $[0,\infty)$ | $\xi,\beta$                   | Teoría de valores extremos.                                                     |
| **Rayleigh**          | Magnitud de vector gaussiano 2D                   |               $[0,\infty)$ | $\sigma$                      | Amplitudes en comunicaciones.                                                   |
| **Wishart**           | Matrices de covarianza muestral                   |            PSD $p\times p$ | $\nu,\Sigma$                  | Suma de productos $XX^\top$.                                                    |

## Relaciones canónicas

* **Poisson ↔ Exponencial/Gamma**: conteos vs tiempos de espera.
  $N(t)\sim\text{Poisson}(\lambda t)$ ⇔ interarribos $X_i\stackrel{iid}{\sim}\text{Exp}(\lambda)$;
  $T_r=\sum_{i=1}^r X_i\sim\text{Gamma}(r,\lambda)$.
* **Bernoulli/Binomial/Geom./NegBin**: familia de ensayos independientes con prob. $p$.
  $\text{Bin}(n,p)=\sum_{i=1}^{n}\text{Bern}(p)$.
  $\text{Geom}(p)=\min\{n:\text{éxito}\}$.
  $\text{NegBin}(r,p)=\sum_{i=1}^{r}\text{Geom}(p)$.
* **Normal, $t$, $\chi^2$, $F$**: inferencia clásica.
  $Z=\frac{\bar X-\mu}{S/\sqrt n}\sim t_{n-1}$ si $X_i\sim\mathcal N$.
  $S^2\propto\chi^2$. Cocientes → $F$.
* **Conjugación bayesiana**:
  $\text{Beta}\leftrightarrow\text{Bern/Bin}$, $\text{Gamma}\leftrightarrow\text{Poisson}$, $\text{Dirichlet}\leftrightarrow\text{Multinomial}$.

## Regla de diseño práctica

1. Identifica el **objeto aleatorio**: conteo, tiempo, proporción, etiqueta, vector.
2. Revisa el **mecanismo**: reemplazo, memoria, suma o producto de choques, colas esperadas.
3. Elige distribución con **soporte y simetrías** correctas.
4. Valida con residuales o pruebas específicas.
5. Si no encaja: transforma, mezcla o usa familias con colas ajustables.

**Atajos de mapeo**

* Conteos raros por intervalo → Poisson $(\lambda)$.
* Tiempos entre eventos Poisson → Exponencial $(\lambda)$.
* Proporciones $[0,1]$ con sesgo → Beta $(\alpha,\beta)$.
* Sumas de muchos choques pequeños → Normal $(\mu,\sigma^2)$.
* Colas pesadas simétricas → $t_\nu$ o Cauchy.
* Fallas con tasa creciente/decreciente → Weibull $(k)$.
* Riqueza, tamaños de cola gruesa → Pareto/GPD.

## Fórmulas mínimas

* Poisson: $\Pr\{X=k\}=\dfrac{e^{-\lambda}\lambda^k}{k!}$.
* Exponencial: $f(x)=\lambda e^{-\lambda x}\,\mathbf 1_{x\ge0}$.
* Binomial: $\Pr\{X=k\}=\binom{n}{k}p^k(1-p)^{n-k}$.
* Normal: $f(x)=\dfrac{1}{\sqrt{2\pi\sigma^2}}\exp\!\left(-\dfrac{(x-\mu)^2}{2\sigma^2}\right)$.
* Beta: $f(x)=\dfrac{x^{\alpha-1}(1-x)^{\beta-1}}{B(\alpha,\beta)}$.
* Weibull: $f(x)=\dfrac{k}{\lambda}\left(\dfrac{x}{\lambda}\right)^{k-1}\!\exp\!\left[-(x/\lambda)^k\right]$.
