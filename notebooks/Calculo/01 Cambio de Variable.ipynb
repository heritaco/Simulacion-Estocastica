# Cambio de variable para $1 - u^2$

## Cálculo analítico

### $\mathbb{E}[U]$

$$
\int_{0}^{1}u\,du=\left[\frac{u^{2}}{2}\right]_{0}^{1}=\frac12.
$$

### $\mathbb{E}\!\left[\sqrt{1-U^{2}}\right]$

Sustitución $u=\sin\theta\Rightarrow du=\cos\theta\,d\theta$, $\theta\in[0,\tfrac{\pi}{2}]$:

$$
\int_{0}^{1}\sqrt{1-u^{2}}\,du=\int_{0}^{\pi/2}\cos^{2}\theta\,d\theta
=\left[\tfrac12\theta+\tfrac14\sin 2\theta\right]_{0}^{\pi/2}=\frac{\pi}{4}.
$$

### $\mathbb{E}\!\left[U\sqrt{1-U^{2}}\right]$

Sustitución $t=1-u^{2}\Rightarrow dt=-2u\,du$:

$$
\int_{0}^{1}u\sqrt{1-u^{2}}\,du=\frac12\int_{0}^{1} t^{1/2}\,dt
=\frac12\left[\frac{2}{3}t^{3/2}\right]_{0}^{1}=\frac13.
$$

### Varianzas

$$
\mathbb{E}[U^{2}]=\int_{0}^{1}u^{2}\,du=\left[\frac{u^{3}}{3}\right]_{0}^{1}=\frac13,
\qquad \operatorname{Var}(U)=\frac13-\left(\frac12\right)^{2}=\frac{1}{12}.
$$

$$
\mathbb{E}\!\left[(\sqrt{1-U^{2}})^{2}\right]=\mathbb{E}[1-U^{2}]=1-\frac13=\frac{2}{3},
$$

$$
\operatorname{Var}\!\left(\sqrt{1-U^{2}}\right)=\frac{2}{3}-\left(\frac{\pi}{4}\right)^{2}.
$$

### Covarianza

$$
\operatorname{Cov}\!\left(U,\sqrt{1-U^{2}}\right)
=\mathbb{E}\!\left[U\sqrt{1-U^{2}}\right]-\mathbb{E}[U]\;\mathbb{E}\!\left[\sqrt{1-U^{2}}\right]
=\frac13-\frac12\cdot\frac{\pi}{4}=\frac13-\frac{\pi}{8}\approx -0.059365748.
$$

### Correlación

$$
\operatorname{Corr}\!\left(U,\sqrt{1-U^{2}}\right)
=\frac{\tfrac13-\tfrac{\pi}{8}}{\sqrt{\tfrac{1}{12}\left(\tfrac{2}{3}-\left(\tfrac{\pi}{4}\right)^{2}\right)}}
\approx -0.921383513.
$$


Sí se puede con $t=1-u^{2}$. Ambas sustituciones sirven. La trigonométrica solo limpia la raíz más rápido.

### Con $t=1-u^{2}$ en $\displaystyle I=\int_{0}^{1}\sqrt{1-u^{2}}\,du$

$$
t=1-u^{2}\;\Rightarrow\;u=\sqrt{1-t},\quad dt=-2u\,du\;\Rightarrow\;du=-\frac{dt}{2u}=-\frac{dt}{2\sqrt{1-t}}.
$$

Límites: $u:0\to1$ implica $t:1\to0$.

$$
I=\int_{1}^{0}\sqrt{t}\left(-\frac{dt}{2\sqrt{1-t}}\right)=\frac12\int_{0}^{1}\frac{t^{1/2}}{(1-t)^{1/2}}\,dt.
$$

Esto es una Beta:

$$
I=\frac12\,B\!\left(\tfrac32,\tfrac12\right)
=\frac12\,\frac{\Gamma(\tfrac32)\Gamma(\tfrac12)}{\Gamma(2)}
=\frac12\cdot\frac{\frac{\sqrt\pi}{2}\cdot\sqrt\pi}{1}
=\frac{\pi}{4}.
$$

### Por qué muchos eligen $u=\sin\theta$

$$
u=\sin\theta\Rightarrow \sqrt{1-u^{2}}=\cos\theta,\quad du=\cos\theta\,d\theta,\quad \theta:0\to\frac{\pi}{2}.
$$

Entonces

$$
I=\int_{0}^{\pi/2}\cos^{2}\theta\,d\theta=\left[\tfrac12\theta+\tfrac14\sin 2\theta\right]_{0}^{\pi/2}=\frac{\pi}{4}.
$$

Ventaja: la raíz desaparece y no aparece un cociente $t^{1/2}/(1-t)^{1/2}$.

### Dónde sí conviene $t=1-u^{2}$

En $\displaystyle \int_{0}^{1}u\sqrt{1-u^{2}}\,du$ es ideal:

$$
t=1-u^{2},\; dt=-2u\,du\;\Rightarrow\;\int_{0}^{1}u\sqrt{1-u^{2}}\,du=\frac12\int_{0}^{1}t^{1/2}\,dt=\frac13.
$$

### Conexión entre ambas

Tomar $u=\sin\theta$ equivale a $t=1-u^{2}=\cos^{2}\theta$. Es la misma familia de cambios. La versión trigonométrica evita la Beta en $\int\sqrt{1-u^{2}}$, pero la de $t$ funciona y es correcta.


Regla práctica:

1. **Si aparece $\sqrt{1-u^{2}}$ sin un factor $u$ que “acompañe” al $du$**
   → Usa $u=\sin\theta$. La raíz desaparece: $\sqrt{1-u^{2}}=\cos\theta$ y $du=\cos\theta\,d\theta$.
   Ej.: $\displaystyle \int \sqrt{1-u^{2}}\,du,\ \int \frac{du}{\sqrt{1-u^{2}}},\ \int \frac{\sqrt{1-u^{2}}}{u}\,du$.

2. **Si hay un factor $u$ fuera de la raíz con potencia impar**
   → Usa $t=1-u^{2}$ porque $dt=-2u\,du$ mata ese $u$.
   Ej.: $\displaystyle \int u\sqrt{1-u^{2}}\,du,\ \int \frac{u}{\sqrt{1-u^{2}}}\,du,\ \int u(1-u^{2})^{p}\,du$.

3. **Potencias generales** $\displaystyle \int u^{m}\,(1-u^{2})^{p}\,du$

* Si $m$ es **impar**, escribe $m=2k+1$ y usa $t=1-u^{2}$: $u\,du=-dt/2$ y $u^{2}=1-t$.
* Si $m$ es **par** o no hay $u$ fuera de la raíz, la trigonométrica $u=\sin\theta$ simplifica.

4. **Integrales definidas en $[0,1]$**

* $u=\sin\theta$ mapea limpio a $\theta\in[0,\pi/2]$ y evita cocientes.
* $t=1-u^{2}$ o $t=u^{2}$ llevan a integrales Beta cuando ves formas $t^{\alpha-1}(1-t)^{\beta-1}$.

5. **Denominadores con potencias negativas de la raíz**
   $(1-u^{2})^{-\alpha}$ suele ser más directo con $u=\sin\theta$ porque aparecen $\sec$ o $\tan$ simples.

---

Ejemplos rápidos:

* $\displaystyle \int_{0}^{1}\sqrt{1-u^{2}}\,du$ → $u=\sin\theta$ ⇒ $\int_{0}^{\pi/2}\cos^{2}\theta\,d\theta=\pi/4$.
* $\displaystyle \int_{0}^{1}u\sqrt{1-u^{2}}\,du$ → $t=1-u^{2}$ ⇒ $\tfrac12\int_{0}^{1}t^{1/2}dt=\tfrac13$.
* $\displaystyle \int \frac{du}{\sqrt{1-u^{2}}}$ → $u=\sin\theta$ ⇒ $\arcsin u + C$.

Resumen: **raíz “sola” o con potencias pares → $u=\sin\theta$**; **$u$ con potencia impar que multiplica → $t=1-u^{2}$**. Generaliza a $\sqrt{a^{2}-u^{2}}$ con $u=a\sin\theta$.
