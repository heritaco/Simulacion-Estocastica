## 1Ô∏è‚É£ Por qu√© $U(m)$ es un grupo

En matem√°ticas, un **grupo** es un conjunto con una operaci√≥n que cumple 4 propiedades:

1. **Cierre:**
   Si $a, b \in U(m)$, entonces $a \cdot b \ (\bmod\ m) \in U(m)$.
