# **QUE ES LA FUNCION TOTIENTE DE EULER?**

La **funci√≥n totiente de Euler**, denotada $\varphi(n)$, es una funci√≥n aritm√©tica que cuenta **cu√°ntos enteros positivos menores o iguales que $n$ son coprimos con $n$** (es decir, tienen m√°ximo com√∫n divisor igual a 1 con $n$).

---

## 1Ô∏è‚É£ Definici√≥n formal

$$
\varphi(n) = \left| \{ k \in \mathbb{Z} \mid 1 \leq k \leq n,\ \gcd(k,n) = 1 \} \right|
$$

Aqu√≠:

* $\gcd(k, n)$ = m√°ximo com√∫n divisor de $k$ y $n$.
* El s√≠mbolo $|\cdot|$ denota el n√∫mero de elementos en el conjunto.

---

## 2Ô∏è‚É£ Ejemplos

* $n=5$ (primo): los n√∫meros $1,2,3,4$ son todos coprimos con 5.
  ‚Üí $\varphi(5) = 4$.
* $n=8$: coprimos con 8 son $1,3,5,7$.
  ‚Üí $\varphi(8) = 4$.
* $n=12$: coprimos con 12 son $1,5,7,11$.
  ‚Üí $\varphi(12) = 4$.

---

## 3Ô∏è‚É£ Propiedades importantes

1. **Si $p$ es primo:**

   $$
   \varphi(p) = p - 1
   $$

   porque todos los enteros menores que $p$ son coprimos con √©l.
2. **Si $n = p^k$ con $p$ primo:**

   $$
   \varphi(p^k) = p^k - p^{k-1} = p^k \left( 1 - \frac{1}{p} \right)
   $$
3. **Si $n$ es factorizable en primos distintos**:

   $$
   \varphi(n) = n \prod_{p\ \text{primo}|n} \left( 1 - \frac{1}{p} \right)
   $$

   donde el producto es sobre los primos que dividen a $n$.

---

## 4Ô∏è‚É£ Relaci√≥n con generadores y periodos

En un **generador congruencial multiplicativo** con m√≥dulo $m$, el tama√±o de $\varphi(m)$ es el **m√°ximo n√∫mero de residuos no nulos que se pueden recorrer** (en el mejor de los casos), porque es exactamente el tama√±o del **grupo multiplicativo** $U(m)$ de enteros invertibles m√≥dulo $m$.

# **Generador congruencial mixto**

Para un **generador congruencial lineal mixto**

$$
x_n \equiv (a x_{n-1} + c) \pmod m
$$

hay reglas matem√°ticas muy claras para escoger $c$ y garantizar **periodo completo**.

---

## 1Ô∏è‚É£ Condiciones de periodo completo (Teorema de Hull‚ÄìDobell)

Un LCG mixto tiene periodo m√°ximo $m$ **si y solo si**:

1. **$\gcd(c, m) = 1$**
   ‚Üí $c$ debe ser **coprimo** con $m$.
   Si $m$ es potencia de 2, esto implica que **$c$ debe ser impar**.
2. **$a - 1$** divisible por todos los primos que dividen $m$.

   * Si $m$ es potencia de 2, basta con que $a \equiv 1 \ (\bmod\ 4)$.
3. **Si $m$ es m√∫ltiplo de 4**, $a - 1$ debe ser m√∫ltiplo de 4.

---

## 2Ô∏è‚É£ C√≥mo escoger $c$

* Primero fijas $m$ (por ejemplo $2^{32}$ o $2^{8}$ como en nuestro ejemplo).
* Buscas un n√∫mero **impar** para $c$ (porque $\gcd(c, m) = 1$ si $m$ es potencia de 2).
* No hace falta que $c$ sea grande; cualquier impar sirve para la condici√≥n 1.
  Ejemplo: con $m = 256$, $c$ podr√≠a ser 1, 3, 5, 7, ..., 255.
* En la pr√°ctica, se elige $c$ que:

  * No cause patrones ‚Äúvisibles‚Äù en las primeras iteraciones.
  * Sea f√°cil de manejar en binario (por ejemplo, constante peque√±a y no m√∫ltiplo de patrones binarios).
  * En generadores hist√≥ricos como *Minimal Standard* o *RANDU* se usan constantes ‚Äúfamosas‚Äù probadas estad√≠sticamente.

---

## 3Ô∏è‚É£ Ejemplo con $m = 256$

* Condici√≥n 1: $c$ impar.
* Condici√≥n 2: $a \equiv 1 \ (\bmod\ 4)$, por ejemplo $a = 5, 9, 13, 17, \dots$.
* As√≠:

  * $a = 13, c = 17$ ‚Üí periodo = 256 ‚úî
  * $a = 5, c = 1$ ‚Üí periodo = 256 ‚úî
  * $a = 9, c = 255$ ‚Üí periodo = 256 ‚úî

---

Para **escoger $c$** en un LCG mixto:

1. Aseg√∫rate que $\gcd(c, m) = 1$ ‚Üí si $m$ es potencia de 2, elige $c$ impar.
